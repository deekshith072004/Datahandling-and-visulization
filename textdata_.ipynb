```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
```

```python
text_data="Python is a popular programming language used for data analysis and  visualization"
```

```python
wordcloud=WordCloud(width=800, height=400, background_color='black').generate(text_data)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.title('World Cloud')
plt.show()
```

```python
words=text_data.split()
```

```python
word_counts=Counter(words)
```

```python
plt.figure(figsize=(10,5))
plt.bar(word_counts.keys(),word_counts.values())
plt.title('Histogram of word Frequencies')
plt.xlabel('words')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```
